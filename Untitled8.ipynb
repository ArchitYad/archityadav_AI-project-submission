```python
import re
import pandas as pd
from nltk.tokenize import word_tokenize


contractions = {
    "can't": "cannot", "wouldn't": "would not", "didn't": "did not", "you're": "you are",
    "i'll": "i will", "it's": "it is", "i'm": "i am", "we're": "we are","hasn't":"has not",
    "i'd": "i had", "doesn't": "does not", "don't": "do not", "what's": "what is",
    "he's": "he is", "she's": "she is", "we'll": "we will", "where's": "where is"
}

def clean_email_text(text):
    if pd.isna(text):
        return ""

    text = text.lower()


    time_preserve = {}

    def save_time(match):
        key = f"time {len(time_preserve)}"
        time_preserve[key] = match.group(0)
        return key

    time_patterns = [
        r'\b\d{1,2}(:\d{2}){0,2}\s*(am|pm)?\s*[-to]+\s*\d{1,2}(:\d{2}){0,2}(am|pm)?\b',
        r'\b\d{1,2}(:\d{2}){1,2}\s*(am|pm)\b'
    ]
    for pattern in time_patterns:
        text = re.sub(pattern, save_time, text, flags=re.IGNORECASE)

    date_preserve = {}

    def save_date(match):
        key = f"date {len(date_preserve)}"
        date_preserve[key] = match.group(0)
        return key

    text = re.sub(r'\b\d{1,2}[/-]\d{1,2}[/-]\d{2,4}\b', save_date, text)
    text = re.sub(r'\b\d{4}[/-]\d{1,2}[/-]\d{1,2}\b', save_date, text)

    month_pattern = r'(?:jan|feb|mar|apr|may|jun|jul|aug|sep|oct|nov|dec|' \
                    r'january|february|march|april|may|june|july|august|' \
                    r'september|october|november|december)'
    date_range_pattern = fr'\b{month_pattern}\s+\d{{1,2}}\s*(–|-|to)\s*\d{{1,2}}\b'
    text = re.sub(date_range_pattern, save_date, text, flags=re.IGNORECASE)



    for c, full in contractions.items():
        text = re.sub(rf"\b{c}\b", full, text)

    text = re.sub(r'\\\\[\w\\]+', '', text)
    text = re.sub(r'\\[\w\\]+', '', text)
    text = re.sub(r'<<.*?>>', '', text)
    text = re.sub(r'>+.*?<+', '', text)
    text = re.sub(r'(\\n)+|\n+|[\r\n\t]+', ' ', text)
    text = re.sub(r'\n[-*~]*\n', ' ', text)

    text = re.sub(r'\[image\]', '', text)
    text = re.sub(r'\b[\w.-]+?@[\w.-]+\.\w{2,4}\b', 'email', text)
    text = re.sub(r'\s@\s', ' at ', text)
    text = re.sub(r'\s@\d', ' at ', text)
    text = re.sub(r'https?://\S+|www\.\S+|mailto:\S+', 'website', text)

    text = re.sub(r'\(?\b\d{3}\)?[-.\s]?\d{3}[-.\s]?\d{4}\b', 'phone number', text)
    text = re.sub(r'\+?\d{1,3}[-.\s]?\d{1,4}[-.\s]?\d{3,5}[-.\s]?\d{4}', 'phone number', text)
    text = re.sub(r'\b[\w\-.]+\.(docx?|pdf|xls[xm]?|jpg|png|exe|pps|txt|zip|html?)\b', 'file', text, flags=re.IGNORECASE)

    text = re.sub(r'\bCorp\.\b', 'Corporation', text)
    text = re.sub(r'\bNat\. Gas\b', 'Natural Gas', text)
    text = re.sub(r'\bNov\.\b', 'November', text)
    text = re.sub(r'\bpg\.\b', 'page', text)

    text = re.sub(r'\$(\d+(?:\.\d+)?)/(\w+)', r'$\1 per \2', text)

    text = re.sub(r'\b(re|fw|fwd)\s*[:\-]', '', text)
    text = re.sub(r'\b(subject|from)\s*[:\-]?', '', text)

    text = re.sub(r'\b\d{6,}\b', '', text)
    text = re.sub(r'(?<!\d),(?!\d)', ' ', text)
    text = re.sub(r'(?<!\d):|:(?!\d)', ' ', text)

    text = re.sub(r'[\[\]\(\)<>=_#*]', ' ', text)
    text = re.sub(r'\.{2,}|\*{2,}|\?{2,}|[-]{2,}|[~]{2,}', ' ', text)
    text = re.sub(r'\/', ' ', text)
    text = re.sub(r'\s*-\s*', ' ', text)

    text = re.sub(r"[’`]", "'", text)
    text = re.sub(r"'\s+", " ", text)
    text = re.sub(r'\s+[\"\,\!\?\;\-]+\s*', ' ', text)

    text = text.replace('"', '')
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'(~\s*)+', '', text)
    text = text.replace("</O=ENRON/OU=NA/CN=RECIPIENTS/CN=DBAUGHM>", "")
    text = re.sub(r'\s*\.\s*', ' ', text)
    text = re.sub(r'@(?=\w)', 'at ', text)

    text = text.strip()

    for key, val in date_preserve.items():
        text = text.replace(key, val)

    for key, val in time_preserve.items():
        text = text.replace(key, val)

    return text.strip()


def remove_repeated_words(text):

    text = re.sub(r'\b(\w+)\s+\1\b', r'\1', text, flags=re.IGNORECASE)

    text = text.strip()

    return text

df.loc[(df['body'].str.strip() == '') & (df['Subject'].str.strip() != ''), 'body'] = df['Subject']

df = df[~((df['Subject'].str.strip() == '') & (df['body'].str.strip() == ''))].reset_index(drop=True)

df['cleaned_text'] = df['body'].apply(clean_email_text).apply(remove_repeated_words)
```

```python
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline

model_name = "cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

sentiment_pipeline = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

def get_sentiment_label(text):
    try:
        result = sentiment_pipeline(text[:512])[0]
        label_map = {
            'LABEL_0': 'Negative',
            'LABEL_1': 'Neutral',
            'LABEL_2': 'Positive'
        }
        return label_map[result['label']]
    except:
        return "Unknown"

df['sentiment'] = df['cleaned_text'].apply(get_sentiment_label)

print(df[['cleaned_text', 'sentiment']].head())
```

```python
def sentiment_score(label):
    if label == 'Positive':
        return 1
    elif label == 'Negative':
        return -1
    else:  # Neutral
        return 0

df['sentiment_score'] = df['sentiment'].apply(sentiment_score)
df['date'] = pd.to_datetime(df['date'])

df['year_month'] = df['date'].dt.to_period('M')
monthly_scores = df.groupby(['from', 'year_month'])['sentiment_score'].sum().reset_index()
print(monthly_scores.head())
```

```python
import pandas as pd
from datetime import timedelta

negative_messages = df[df['sentiment'] == 'Negative'].copy()
negative_messages['date'] = pd.to_datetime(negative_messages['date'])

at_risk_employees = []

for employee in negative_messages['from'].unique():
    employee_messages = negative_messages[negative_messages['from'] == employee]
    employee_messages = employee_messages.sort_values(by='date')

    for i in range(len(employee_messages)):
        current_date = employee_messages.iloc[i]['date']
        window_start = current_date - timedelta(days=30)

        messages_in_window = employee_messages[(employee_messages['date'] >= window_start) &
                                               (employee_messages['date'] <= current_date)]
        if len(messages_in_window) >= 4:
            at_risk_employees.append(employee)
            break

at_risk_employees = list(set(at_risk_employees))

at_risk_df = pd.DataFrame(at_risk_employees, columns=['Employee_Email'])
print(at_risk_df)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

sentiment_mapping = {'Positive': 1, 'Neutral': 0, 'Negative': -1}
df['sentiment'] = df['sentiment'].map(sentiment_mapping)

X = df['cleaned_text']
y = df['sentiment']

vectorizer = TfidfVectorizer(max_features=1000, stop_words='english')
X_vec = vectorizer.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_vec, y, test_size=0.2, random_state=42)

clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objs as go
from plotly.subplots import make_subplots
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error

df['date'] = pd.to_datetime(df['date'])
df = df.sort_values('date').set_index('date')

ts = df['sentiment_score'].resample('D').mean().fillna(0)

train = ts[:-7]
test = ts[-7:]

model = ARIMA(train, order=(5,1,0))
model_fit = model.fit()

forecast = model_fit.forecast(steps=7)

mse = mean_squared_error(test, forecast)
print(f"Mean Squared Error on last 7 days: {mse:.4f}")

dates = ts.index
forecast_dates = pd.date_range(start=dates[-1] + pd.Timedelta(days=1), periods=7)

fig = make_subplots(rows=1, cols=1)

fig.add_trace(go.Scatter(x=dates, y=ts, mode='lines', name='Historical Sentiment'))

fig.add_trace(go.Scatter(x=forecast_dates, y=forecast, mode='lines+markers', name='7-day Forecast'))

fig.update_layout(title="ARIMA Sentiment Score Forecast",
                  xaxis_title="Date",
                  yaxis_title="Sentiment Score",
                  template='plotly_white')

fig.show()
```
